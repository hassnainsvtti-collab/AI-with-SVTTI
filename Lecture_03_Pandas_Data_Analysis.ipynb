## 1. What is Pandas?

**Pandas** is a Python library for working with *structured data* — the kind stored in tables (rows and columns), like an Excel spreadsheet or a database table.

In AI projects you spend **60–80 % of your time on data** — loading, cleaning, and transforming it before any model ever sees it. Pandas is your main tool for all of that.

### Core objects
| Object | Description | Analogy |
|--------|-------------|---------|
| **Series** | 1-D labelled array | A single column in a spreadsheet |
| **DataFrame** | 2-D table (rows × columns) | A full spreadsheet or database table |

```python
# ── CODE CELL 1 ──────────────────────────────────────────────────────────────
# Series and DataFrame basics

import pandas as pd
import numpy as np

# ── Series ───────────────────────────────────────────────────────────────────
marks = pd.Series([85, 90, 78, 92, 88],
                  index=["Ahmed", "Fatima", "Usman", "Sara", "Ali"])
print("Series — Student Marks:")
print(marks)
print(f"\nMax: {marks.max()}  |  Min: {marks.min()}  |  Mean: {marks.mean():.1f}")
print()

# ── DataFrame ────────────────────────────────────────────────────────────────
data = {
    "name":          ["Ahmed", "Fatima", "Usman", "Sara", "Ali", "Zara"],
    "age":           [22, 20, 25, 21, 23, 19],
    "city":          ["Gujranwala", "Lahore", "Karachi", "Islamabad", "Peshawar", "Lahore"],
    "study_hours":   [6, 8, 4, 7, 5, 9],
    "marks":         [85, 92, 65, 88, 72, 96],
    "passed":        [True, True, False, True, True, True],
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)
```

```python
# ── CODE CELL 2 ──────────────────────────────────────────────────────────────
# Exploring a DataFrame — the first thing you do with any new dataset

print("Shape (rows, columns):", df.shape)
print()
print("Column names:", list(df.columns))
print()
print("Data types:")
print(df.dtypes)
print()
print("First 3 rows:")
print(df.head(3))
print()
print("Statistical summary:")
print(df.describe())
```

```python
# ── CODE CELL 3 ──────────────────────────────────────────────────────────────
# Selecting columns and rows

# Select a single column → returns a Series
print("Marks column:")
print(df["marks"])
print()

# Select multiple columns → returns a DataFrame
print("Name and marks:")
print(df[["name", "marks"]])
print()

# Select rows by position — iloc (integer location)
print("First 2 rows (iloc):")
print(df.iloc[0:2])
print()

# Select rows by label — loc
print("Row at index 3 (loc):")
print(df.loc[3])
```

```python
# ── CODE CELL 4 ──────────────────────────────────────────────────────────────
# Filtering rows — boolean conditions

# Students who passed
passed_students = df[df["passed"] == True]
print("Students who passed:")
print(passed_students[["name", "marks", "passed"]])
print()

# Students from Lahore with marks above 80
lahore_high = df[(df["city"] == "Lahore") & (df["marks"] > 80)]
print("Lahore students with marks > 80:")
print(lahore_high[["name", "city", "marks"]])
print()

# Students who studied more than 6 hours
hard_workers = df[df["study_hours"] > 6].sort_values("marks", ascending=False)
print("Hard workers (study_hours > 6), sorted by marks:")
print(hard_workers[["name", "study_hours", "marks"]])
```

```python
# ── CODE CELL 5 ──────────────────────────────────────────────────────────────
# Adding new columns and applying functions

# Add a grade column derived from marks
def assign_grade(mark):
    if   mark >= 90: return "A+"
    elif mark >= 80: return "A"
    elif mark >= 70: return "B"
    elif mark >= 60: return "C"
    else:            return "D"

df["grade"] = df["marks"].apply(assign_grade)

# Add a marks-per-study-hour efficiency column
df["efficiency"] = (df["marks"] / df["study_hours"]).round(1)

print(df[["name", "study_hours", "marks", "grade", "efficiency"]])
```

```python
# ── CODE CELL 6 ──────────────────────────────────────────────────────────────
# Handling missing data — critical in real-world AI projects

# Create a DataFrame with missing values (NaN)
messy_data = pd.DataFrame({
    "name":          ["Ahmed", "Fatima", None,    "Sara",  "Ali",   "Zara"],
    "age":           [22,      20,       25,       None,    23,      19],
    "marks":         [85,      None,     65,       88,      72,      None],
    "city":          ["GRW",   "LHR",    "KHI",   "ISL",   "LHR",   "LHR"],
})

print("Messy DataFrame:")
print(messy_data)
print()
print("Missing values per column:")
print(messy_data.isnull().sum())
print()

# Strategy 1: Fill missing marks with the column mean
messy_data["marks"].fillna(messy_data["marks"].mean(), inplace=True)

# Strategy 2: Fill missing age with median
messy_data["age"].fillna(messy_data["age"].median(), inplace=True)

# Strategy 3: Drop rows where name is missing
messy_data.dropna(subset=["name"], inplace=True)

print("Cleaned DataFrame:")
print(messy_data)
print()
print("Missing values after cleaning:")
print(messy_data.isnull().sum())
```

```python
# ── CODE CELL 7 ──────────────────────────────────────────────────────────────
# Grouping and aggregation — finding patterns by category

print("Average marks by city:")
print(df.groupby("city")["marks"].mean().sort_values(ascending=False))
print()

print("Full stats by grade:")
print(df.groupby("grade").agg(
    student_count=("name", "count"),
    avg_marks=("marks", "mean"),
    avg_study_hours=("study_hours", "mean")
).round(1))
print()

print("Pass rate by city:")
print(df.groupby("city")["passed"].mean().apply(lambda x: f"{x*100:.0f}%"))
```

```python
# ── CODE CELL 8 ──────────────────────────────────────────────────────────────
# Reading and writing CSV files

# Save our DataFrame to a CSV file
df.to_csv("/content/students.csv", index=False)
print("Saved to students.csv")

# Read it back
df_loaded = pd.read_csv("/content/students.csv")
print("\nLoaded from CSV:")
print(df_loaded.head())

# In real projects you will load public datasets like this:
# df = pd.read_csv("https://some-url.com/dataset.csv")
# Or a local file:
# df = pd.read_csv("/content/my_dataset.csv")

print("\nLoading a real-world dataset from the web:")
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
try:
    titanic = pd.read_csv(url)
    print(f"Titanic dataset loaded: {titanic.shape[0]} rows, {titanic.shape[1]} columns")
    print(titanic.head(3))
except Exception as e:
    print(f"Could not load from web: {e}")
    print("This is normal if there is no internet connection in the current session.")
```

```python
# ── CODE CELL 9 ──────────────────────────────────────────────────────────────
# Preparing data for a Machine Learning model
# This is exactly what you will do in Module 3!

print("Original DataFrame:")
print(df[["name", "study_hours", "marks", "passed"]])
print()

# Step 1: Select features (X) and target (y)
X = df[["study_hours", "age"]].values    # .values converts to NumPy array
y = df["marks"].values

print("Feature matrix X (study_hours, age):")
print(X)
print()
print("Target vector y (marks):")
print(y)

# Step 2: Normalise X to [0, 1]
X_norm = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
print("\nNormalised features:")
print(X_norm.round(3))
print()
print("Data is now ready for a Machine Learning model!")
```

## Lecture 3 — Summary

| Operation | Code Example |
|-----------|--------------|
| Create DataFrame | `pd.DataFrame(dict)` |
| Load CSV | `pd.read_csv('file.csv')` |
| Explore | `.head()`, `.shape`, `.describe()`, `.dtypes` |
| Select column | `df['col']` or `df[['col1','col2']]` |
| Filter rows | `df[df['col'] > 50]` |
| Add column | `df['new'] = df['a'] + df['b']` |
| Apply function | `df['col'].apply(func)` |
| Handle NaN | `.fillna()`, `.dropna()`, `.isnull().sum()` |
| Group & aggregate | `df.groupby('col').mean()` |
| Sort | `df.sort_values('col', ascending=False)` |
