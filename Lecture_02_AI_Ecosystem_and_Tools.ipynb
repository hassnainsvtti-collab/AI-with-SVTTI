## 📌 4. Notebook Cells Ka Concept

Jupyter/Colab notebook mein do tarah ki cells hoti hain:

### 📝 Markdown Cell (Text Cell)
- Notes, headings, explanations likhne ke liye
- Bilkul is cell jaisi!
- `#` se heading, `**bold**`, `*italic*`

### 💻 Code Cell
- Python code likhne ke liye
- ▶️ button ya `Shift+Enter` se chalao
- Output neeche dikhta hai

> **Shortcut yaad rakhein:**
> - `Shift + Enter` → Cell chalao aur agli pe jao
> - `Ctrl + Enter` → Sirf yeh cell chalao
> - `Ctrl + M + B` → Neeche naya cell banao

````python
# 🖥️ Python ke Basic Concepts — Quick Overview

# 1. Variables (Data store karna)
naam = "Ahmed"          # Text (String)
umar = 20               # Number (Integer)
marks = 85.5            # Decimal (Float)
pass_hua = True         # Haan/Na (Boolean)

print("Naam:", naam)
print("Umar:", umar)
print("Marks:", marks)
print("Pass hua?:", pass_hua)

print()

# 2. Simple Calculation
a = 10
b = 3
print("Jama (Addition):", a + b)
print("Tafreek (Subtraction):", a - b)
print("Zarb (Multiplication):", a * b)
print("Taqseem (Division):", a / b)
print("Baqi (Remainder):", a % b)
````
