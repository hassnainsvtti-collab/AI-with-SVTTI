## 📌 1. Python Quick Recap — Core Concepts

Module 2 mein Python detail mein seekhein ge. Abhi yeh essentials yaad karein:

````python
# 📝 Lists — Data ki collection

ai_modules = [
    "Module 1: Intro to AI",
    "Module 2: Python for AI",
    "Module 3: Machine Learning",
    "Module 4: Deep Learning",
    "Module 5: Computer Vision"
]

# List ke saath kaam:
print(f"📚 Kul modules: {len(ai_modules)}")
print(f"🥇 Pehla module: {ai_modules[0]}")
print(f"🏁 Akhri module: {ai_modules[-1]}")

# Loop ke saath:
print("\n📋 Tamam Modules:")
for i, module in enumerate(ai_modules, 1):
    print(f"  {i}. {module}")
````

````python
# 📝 Dictionaries — Key-Value data (like a real dictionary)

student = {
    "naam": "Ali Ahmed",
    "umar": 22,
    "module": "Module 1",
    "marks": {"quiz1": 85, "quiz2": 90, "assignment": 88}
}

print("Student Info:")
print(f"  Naam: {student['naam']}")
print(f"  Umar: {student['umar']} saal")
print(f"  Module: {student['module']}")

# Marks calculate karein:
avg_marks = sum(student['marks'].values()) / len(student['marks'])
print(f"  Average Marks: {avg_marks:.1f}%")

grade = "A" if avg_marks >= 90 else "B" if avg_marks >= 80 else "C"
print(f"  Grade: {grade}")
````

````python
# 📝 Functions — Reusable code blocks

def calculate_grade(marks_list):
    """
    Marks ki list le ke grade return karti hai.
    Input: [85, 90, 78]
    Output: Average aur Grade
    """
    average = sum(marks_list) / len(marks_list)

    if average >= 90:
        grade = "A+ (Excellent!)"
    elif average >= 80:
        grade = "A (Very Good)"
    elif average >= 70:
        grade = "B (Good)"
    elif average >= 60:
        grade = "C (Average)"
    else:
        grade = "D (Needs Improvement)"

    return average, grade

# Function use karein:
students_data = {
    "Ahmed": [85, 92, 78, 88],
    "Fatima": [95, 97, 92, 98],
    "Usman": [65, 70, 68, 72],
}

print("📊 Class Results:")
print("=" * 45)
for naam, marks in students_data.items():
    avg, grade = calculate_grade(marks)
    print(f"  {naam:<10} → Avg: {avg:.1f}% | Grade: {grade}")
````

## 📌 2. Python Control Flow — If/Loops

AI models ke andar bhi yeh hi logic hota hai!

````python
# 📝 If/Elif/Else + Loops

# Fibonacci sequence — AI mein patterns dhundha jata hai
def fibonacci(n):
    """Pehle n Fibonacci numbers return karo"""
    if n <= 0:
        return []
    elif n == 1:
        return [0]

    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])  # Pichle do ka jama
    return fib

print("Fibonacci Sequence (pehle 10):")
result = fibonacci(10)
print(result)

print("\n🌀 Yeh pattern AI mein kaafi use hota hai!")

# List Comprehension — Python ka shortcut
print("\nList Comprehension example:")
squares = [x**2 for x in range(1, 11)]
print(f"1 se 10 ke squares: {squares}")

even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print(f"Sirf juft (even) ke squares: {even_squares}")
````

## 📌 3. Mini Project — AI Knowledge Quiz! 🎯

Is module ka final project: Ek **interactive quiz** banayein AI basics ka.

Yeh project sab kuch combine karta hai:

````python
# 🎯 MINI PROJECT: AI Knowledge Quiz System
# Yeh Module 1 ka saara content cover karta hai

quiz_questions = [
    {
        "sawal": "AI ka pura naam kya hai?",
        "options": ["A) Automated Intelligence", "B) Artificial Intelligence",
                    "C) Advanced Information", "D) Automated Information"],
        "jawab": "B",
        "explanation": "AI = Artificial Intelligence — machine ko insaan jaisi soch dena"
    },
    {
        "sawal": "Python AI ke liye kyun best hai?",
        "options": ["A) Sirf Python free hai", "B) Python sabse puraani hai",
                    "C) Easy syntax aur bohat AI libraries", "D) Python sirf AI ke liye hai"],
        "jawab": "C",
        "explanation": "Python easy hai aur TensorFlow, PyTorch, Scikit-learn jaisi libraries hain"
    },
    {
        "sawal": "Google Colab kya hai?",
        "options": ["A) Google ka AI model", "B) Online Python notebook — free",
                    "C) Python install karne ka tareeqa", "D) GitHub ki tarah"],
        "jawab": "B",
        "explanation": "Colab browser mein chalta hai — kuch install nahi karna"
    },
    {
        "sawal": "Aaj kal ki AI kis type ki hai?",
        "options": ["A) General AI (AGI)", "B) Super AI",
                    "C) Narrow AI", "D) Robot AI"],
        "jawab": "C",
        "explanation": "Narrow AI sirf ek kaam karta hai — lekin woh kaam bahut acha karta hai"
    },
    {
        "sawal": "GitHub ky kaam aata hai?",
        "options": ["A) AI models train karna", "B) Code safe rakhna aur share karna",
                    "C) Python install karna", "D) Data store karna"],
        "jawab": "B",
        "explanation": "GitHub code ka ghar hai — portfolio aur collaboration ke liye"
    }
]

def quiz_chalao(questions):
    """Quiz run karne ka function"""
    score = 0
    total = len(questions)

    print("=" * 55)
    print("🎯 DEV VALLEY AI KNOWLEDGE QUIZ — MODULE 1")
    print("=" * 55)
    print(f"Total Sawaalat: {total}")
    print("Ek letter type karein (A/B/C/D) aur Enter dabayein")
    print("=" * 55)

    for i, q in enumerate(questions, 1):
        print(f"\n❓ Sawal {i}/{total}: {q['sawal']}")
        for option in q['options']:
            print(f"   {option}")

        jawab = input("\nAapka jawab: ").upper().strip()

        if jawab == q['jawab']:
            print("   ✅ Sahi! Bilkul theek!")
            score += 1
        else:
            print(f"   ❌ Ghalat. Sahi jawab: {q['jawab']}")

        print(f"   💡 Explanation: {q['explanation']}")

    # Final result
    print("\n" + "=" * 55)
    print(f"🏆 FINAL SCORE: {score}/{total} ({score/total*100:.0f}%)")

    if score == total:
        print("🌟 PERFECT SCORE! Ustaz ke layiq student! 👏")
    elif score >= total * 0.8:
        print("🎉 Bohat acha! AI Engineer banne ki raah pe ho!")
    elif score >= total * 0.6:
        print("👍 Theek hai! Thodi mazeed practice karein.")
    else:
        print("📚 Lectures dobara sun lo — koi baat nahi, practice se aayega!")
    print("=" * 55)

# Quiz chalao!
quiz_chalao(quiz_questions)
````
