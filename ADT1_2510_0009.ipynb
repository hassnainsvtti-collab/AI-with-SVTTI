```python
import pandas as pd
```

```python
df=pd.read_csv('/content/home_prices.csv')
```

```python
df.head()
```

```python
X=df[['area','bedrooms','age']]
Y=df['price']
```

```python
from sklearn import linear_model
import pandas as pd # Import pandas for data manipulation

# Combine X and Y into a single DataFrame to drop rows with NaNs consistently
combined_data = pd.concat([X, Y], axis=1)
cleaned_data = combined_data.dropna()

# Separate X_cleaned and Y_cleaned
X_cleaned = cleaned_data[X.columns]
Y_cleaned = cleaned_data[Y.name]

model=linear_model.LinearRegression()
model.fit(X_cleaned,Y_cleaned) # Fit the model with cleaned data

# Make prediction, providing input as a DataFrame with column names
prediction_input = pd.DataFrame([[2500,4,5]], columns=X.columns)
model.predict(prediction_input)
```
