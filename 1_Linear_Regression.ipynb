```python
import pandas as pd
```

```python
df=pd.read_csv ('/content/homeprices.csv')
```

```python
df.head(8)
```

```python
X=df[['area']]
```

```python
Y=df['price']
```

```python
import matplotlib.pyplot as plt
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(X,Y,color='red',marker='+')
```

```python
# Create linear regression object
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X,Y)
```

**(1) Predict price of a home with area=90000 sqr ft**

Add blockquote

```python
model.predict([[90000]])
```

```python
model.coef_
```
