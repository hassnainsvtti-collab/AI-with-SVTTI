## 📌 1. GitHub Kya Hai?

**GitHub = Code ka Google Drive** — lekin bahut zyada powerful!

### Kyun Zaruri Hai?
```
❌ Bina GitHub ke:
   - Laptop kharab ho gayi → sab code gaya
   - Kisi ko share karna mushkil
   - Purana version wapas nahi aata

✅ GitHub ke saath:
   - Code safe hai cloud mein
   - Koi bhi dekh sakta hai (portfolio!)
   - Purana version wapas la sakte hain
   - Team ke saath kaam kar sakte hain
```

### Real Life Analogy:
Sochein ke aap ek book likh rahe hain:
- **Git** = History rakhna (draft 1, draft 2, draft 3...)
- **GitHub** = Woh book online publish karna

## 📌 4. Git Ki Vocabulary — Samjhein

| Git Term | Asaan Urdu Matlab |
|----------|------------------|
| **Repository (Repo)** | Project ka folder |
| **Commit** | Changes save karna (snapshot lena) |
| **Push** | Local se GitHub pe bhejna |
| **Pull** | GitHub se apne computer pe lana |
| **Clone** | Kisi ki repo copy karna |
| **Branch** | Alag version mein kaam karna |
| **Merge** | Do branches ko jodna |
| **README.md** | Project ki description file |

### Aksar Istemaal Hone Wale Commands:
```bash
git clone <url>        # Repo download karo
git add .              # Sab changes stage karo
git commit -m "message" # Snapshot lo
git push               # GitHub pe bhejo
git pull               # GitHub se naya version lao
git status             # Kya changes hain dekho
git log                # History dekho
```

````python
# 🖥️ Git Commands Demo — Colab mein

import os

# Ek demo folder banao aur git initialize karo
!mkdir -p /content/demo_project
os.chdir("/content/demo_project")

# Git initialize
!git init

# Ek file banao
with open("hello.py", "w") as f:
    f.write('print("Mera pehla AI project!")')

with open("README.md", "w") as f:
    f.write("# Mera AI Project\nDev Valley Gujranwala")

# Git Status check
print("\n📋 Git Status:")
!git status
````

````python
# 📸 Commit karna — Snapshot lena

!git add .
!git commit -m "First commit: AI course shuru! 🚀"

print("\n📜 Git Log (History):")
!git log --oneline
````
